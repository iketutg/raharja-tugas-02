```python
##Constants Variabel 
PROD_PAY = "PAYMENT" 
PROD_TRF = "TRANSFER" 
PROD_CASHIN = "CASHIN" 
PROD_CASHOUT = "CASHOUT" 
PROD_INQBAL ="INQBALANCE" 
PROD_SETTLEMENT = "SETTLEMENT"

KEY_FREQ_LIMIT_MINUTES = "KEYFREQLIMITMIN" 
KEY_AMT_LIMIT_MINUTES = "KEYAMTLIMITMIN" 

KEY_FREQ_LIMIT_DAY = "KEYFREQLIMITDAY" 
KEY_AMT_LIMIT_DAY = "KEYAMTLIMITDAY" 

KEY_FREQ_LIMIT_MONTH = "KEYFREQLIMITMONTH" 
KEY_AMT_LIMIT_MONTH = "KEYAMTLIMITMONTH" 

KEY_FREQ_LIMIT_YEAR = "KEYFREQLIMITYEAR" 
KEY_AMT_LIMIT_YEAR = "KEYAMTLIMITYEAR" 

KEY_INPUT_PRODUCT = "product" 
KEY_INPUT_FREQ_MIN = "freqmin" 
KEY_INPUT_AMT_MIN = "amtmin"
KEY_INPUT_FREQ_DAY = "freqday" 
KEY_INPUT_AMT_DAY = "amtday"

KEY_INPUT_FREQ = "freq" 
KEY_INPUT_AMT = "amt"

KEY_MINUTES = "MINUTES"
KEY_DAY = "DAY"
KEY_MONTH = "MONTH" 
KEY_YEAR = "YEAR"
# selectproduct = {1:PROD_PAY,2:PROD_TRF,3:PROD_CASHIN,4:PROD_CASHOUT,5:PROD_INQBAL,6:PROD_SETTLEMENT}
## global variabel data mahasiswa 
listProdFilter = (PROD_PAY,PROD_TRF,PROD_CASHIN,PROD_CASHOUT) # tuple

cfgfreqlimit = {KEY_FREQ_LIMIT_MINUTES:3,KEY_FREQ_LIMIT_DAY : 5 , 
                KEY_FREQ_LIMIT_MONTH : 7, KEY_FREQ_LIMIT_YEAR : 9}

cfgamtlimit = {KEY_AMT_LIMIT_MINUTES:11,KEY_AMT_LIMIT_DAY:41 , 
               KEY_AMT_LIMIT_MONTH : 61 , KEY_AMT_LIMIT_YEAR : 81} 

##Data 
dataminutes = {KEY_INPUT_FREQ_MIN:0,KEY_INPUT_AMT_MIN:0} 
datadaily = {KEY_INPUT_FREQ_DAY:0,KEY_INPUT_AMT_DAY:0} 
datamonth = {KEY_INPUT_FREQ:0,KEY_INPUT_AMT:0}
datayear = {KEY_INPUT_FREQ:0,KEY_INPUT_AMT:0}
dbmemory = {KEY_MINUTES:dataminutes,KEY_DAY:datadaily,KEY_MONTH:datamonth, KEY_YEAR:datayear}

inputdaily ={KEY_INPUT_PRODUCT:"",KEY_INPUT_FREQ:0,KEY_INPUT_AMT:0}
```

```python
# func pengecekan product filter
def rule_product_needFilter(product): 
    return product in listProdFilter
```

```python
################################## 
## func limit in minutes 
def rule_freq_limit_minutes(freq): 
    dbmin = dbmemory.get(KEY_MINUTES)
    freq_min = dbmin.get(KEY_INPUT_FREQ_MIN) 
    freq += freq_min 
    if freq > cfgfreqlimit[KEY_FREQ_LIMIT_MINUTES]: 
        return True
    else: 
        return False 

def rule_amt_limit_minutes(amt): 
    dbmin = dbmemory.get(KEY_MINUTES)
    amt_min = dbmin.get(KEY_INPUT_AMT_MIN) 
    amt += amt_min 
    if amt > cfgamtlimit[KEY_AMT_LIMIT_MINUTES]: 
        return True
    else: 
        return False 

def rule_limit_minutes(freq,amt): 
    if rule_freq_limit_minutes(freq) or rule_amt_limit_minutes(amt): 
        return True 
    else: 
        return False 
    
```

```python
##################################    
## func limit in days     
def rule_freq_limit_days(freq): 
    dbdaily = dbmemory.get(KEY_DAY)
    freq_day = dbdaily.get(KEY_INPUT_FREQ_DAY)
    freq += freq_day 
    
    if freq > cfgfreqlimit[KEY_FREQ_LIMIT_DAY]: 
        return True
    else: 
        return False 

def rule_amt_limit_days(amt): 
    dbdaily = dbmemory.get(KEY_DAY)
    amt_day = dbdaily.get(KEY_INPUT_AMT_DAY)
    amt += amt_day 
    if amt > cfgamtlimit[KEY_AMT_LIMIT_DAY]: 
        return True
    else: 
        return False 
    

def rule_limit_days(freq,amt):
    if rule_freq_limit_days(freq) or rule_amt_limit_days(amt): 
        return True 
    else: 
        return False 
    
```

```python
##################################
## func limit in days    
def rule_freq_limit_month(freq):
    dbmonth = dbmemory.get(KEY_MONTH)
    freq_month = dbmonth.get(KEY_INPUT_FREQ)
    freq += freq_month
    
    if freq > cfgfreqlimit[KEY_FREQ_LIMIT_MONTH]: 
        return True
    else: 
        return False 

def rule_amt_limit_month(amt): 
    dbmonth = dbmemory.get(KEY_MONTH)
    amt_month = dbmonth.get(KEY_INPUT_AMT)
    amt += amt_month
    
    if amt > cfgamtlimit[KEY_AMT_LIMIT_MONTH]: 
        return True
    else: 
        return False     
    
def rule_limit_month(freq,amt): 
    if rule_freq_limit_month(freq) or rule_amt_limit_month(amt): 
        return True 
    else: 
        return False     
    
```

```python
##################################
## func limit in year    
def rule_freq_limit_year(freq): 
    dbyear = dbmemory.get(KEY_YEAR)
    freq_year = dbyear.get(KEY_INPUT_FREQ)
    freq += freq_year
    
    if freq > cfgfreqlimit[KEY_FREQ_LIMIT_YEAR]: 
        return True
    else: 
        return False 

def rule_amt_limit_year(amt): 
    dbyear = dbmemory.get(KEY_YEAR)
    amt_year = dbyear.get(KEY_INPUT_AMT)
    amt += amt_year
    
    if amt > cfgamtlimit[KEY_AMT_LIMIT_YEAR]: 
        return True
    else: 
        return False     
    
def rule_limit_year(freq,amt): 
    if rule_freq_limit_year(freq) or rule_amt_limit_year(amt): 
        return True 
    else: 
        return False   
    
```

```python
##################################
def send_notif_alert(message): 
    print("######################")
    print("STOP")
    print(message)
    print("######################")
    
def next_porcess(): 
    print("===================")
    print("OK PROCESS SUCCESS")
    print("===================")
```

```python
##################################
# function save data in memory 

def save_trans_minutes(inputdaily):
    dbminutes = dbmemory.get(KEY_MINUTES)
    ##dataminutes = {KEY_INPUT_FREQ_MIN:0,KEY_INPUT_AMT_MIN:0} 
    dbminutes[KEY_INPUT_FREQ_MIN] += inputdaily[KEY_INPUT_FREQ]
    dbminutes[KEY_INPUT_AMT_MIN] += inputdaily[KEY_INPUT_AMT]
    dbmemory[KEY_MINUTES] = dbminutes

    
def save_trans_daily(inputdaily):
    dbdaily = dbmemory.get(KEY_DAY)
    dbdaily[KEY_INPUT_FREQ_DAY] += inputdaily[KEY_INPUT_FREQ]
    dbdaily[KEY_INPUT_AMT_DAY] += inputdaily[KEY_INPUT_AMT]
    dbmemory[KEY_DAY] = dbdaily

    
def save_trans_monthly(inputdaily):
    dbmonth = dbmemory.get(KEY_MONTH)
    dbmonth[KEY_INPUT_FREQ] += inputdaily[KEY_INPUT_FREQ]
    dbmonth[KEY_INPUT_AMT] += inputdaily[KEY_INPUT_AMT]
    dbmemory[KEY_MONTH] = dbmonth

def save_trans_yearly(inputdaily):
    dbyear = dbmemory.get(KEY_YEAR)
    dbyear[KEY_INPUT_FREQ] += inputdaily[KEY_INPUT_FREQ]
    dbyear[KEY_INPUT_AMT] += inputdaily[KEY_INPUT_AMT]
    dbmemory[KEY_YEAR] = dbyear
    
    
def save_transaction(inputdaily): 
    save_trans_minutes(inputdaily)
    save_trans_daily(inputdaily)
    save_trans_monthly(inputdaily)
    save_trans_yearly(inputdaily)
    print("Data Save {} ".format(dbmemory))
```

```python
##################################
## func reset memory 

def reset_minutes(): 
    dataminutes = {KEY_INPUT_FREQ_MIN:0,KEY_INPUT_AMT_MIN:0} 
    dbmemory[KEY_MINUTES]  = dataminutes
    
def reset_days(): 
    datadaily = {KEY_INPUT_FREQ_DAY:0,KEY_INPUT_AMT_DAY:0} 
    dbmemory[KEY_DAY]  = datadaily
    
def reset_month(): 
    datamonth = {KEY_INPUT_FREQ:0,KEY_INPUT_AMT:0} 
    dbmemory[KEY_MONTH]  = datamonth
##################################
```

```python
################################## 
## tree logical 
def func_logical_limitation(inputdaily): 
    if rule_product_needFilter(inputdaily['product']): 
        ##TODO condition
        #freq = datainput[KEY_INPUT_FREQ]
        # amt = datainput[KEY_INPUT_AMT]
        if rule_limit_minutes(inputdaily[KEY_INPUT_FREQ],inputdaily[KEY_INPUT_AMT]): 
                send_notif_alert("TRANSACTION MELEBIH LIMIT (MINUTES)")
                return True
        else:
             if rule_limit_days(inputdaily[KEY_INPUT_FREQ],inputdaily[KEY_INPUT_AMT]): 
                send_notif_alert("TRANSACTION MELEBIH LIMIT (DAYS)")
                return True 
             else:   
                if rule_limit_month(inputdaily[KEY_INPUT_FREQ],inputdaily[KEY_INPUT_AMT]): 
                    send_notif_alert("TRANSACTION MELEBIH LIMIT (MONTH)")
                    return True 
                else:
                    if rule_limit_year(inputdaily[KEY_INPUT_FREQ],inputdaily[KEY_INPUT_AMT]): 
                        send_notif_alert("TRANSACTION MELEBIH LIMIT (YEAR)")
                        return True 
                    else:
                        save_transaction(inputdaily)
                        next_porcess()
                        return False 
    else : 
        print("This Product No Need Filter")
        next_porcess()
        return False  
```
